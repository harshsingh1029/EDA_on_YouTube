### 6. Create a new column as "published_month" using the column published_at (display the months only)

```python
def extract_month(data,new_col,col):# Extract the month from the 'published_at' column and create a new column 'published_month'
    data[new_col]=data[col].dt.month
    print("Extraction successfully done...")
    print(pd.DataFrame(data[new_col].head()))
    return data
```

```python
df=extract_month(df,new_col="published_month",col="published_at" )#Calling Function
```

```python
pd.DataFrame(df["published_month"])#Checking new column published_month
```

<font size=4><font color="green">=>Create new column <font color="orange">df["published_month"]<font color="green"> and assign the months from column <font color="orange">df["published_at"]<font color="green"> by using <font color="orange">.dt.month

### 7. Replace the numbers in the column published_month as names of the months i,e., 1 as "Jan", 2 as "Feb"
and so on.....

```python
def month_name(data,col):# Function Convert month number into month name
    data[col]=data[col].apply(lambda x: calendar.month_name[x])
    print("Convert successfully...")
    print(pd.DataFrame(data[col].head()))
    return data
    
```

```python
df=month_name(df,col="published_month")
```

```python
pd.DataFrame(df["published_month"])#Show changed in Column
```

```python
def log_df(data):# Function to Log normalization
 for col in data.select_dtypes([int,float]).columns:
  data[col]=np.log1p(data[col])
 return data
```
